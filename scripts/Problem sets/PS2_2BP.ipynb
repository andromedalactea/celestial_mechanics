**2.** **Traslación de las cónicas.** A partir de una completación de cuadrados, demuestre que la ecuación general de las cónicas sin términos acoplados

$$Ax^2+Cy^2+Dx+Ey+G=0$$

con $A,C\neq 0$ se puede escribir, en su forma ordinaria, como

$$\frac{x'^2}{q}+\frac{y'^2}{t}=1,$$

donde $x'$ y $y'$ son las coordenadas de un punto en el sistema trasladado dadas por

$$x' = x+\frac{D}{2A},$$
$$y' = y+\frac{E}{2C}$$

y $q,t$ están dadas por

$$q	= -\frac{G}{A}+\frac{D^{2}}{4A^{2}}+\frac{E^{2}}{4AC},$$
$$t	= -\frac{G}{C}+\frac{D^{2}}{4AC}+\frac{E^{2}}{4C^{2}}.$$

¿Qué condiciones se deben dar para que dicha cónica sea una circunferencia, una elipse o una hipérbola? Dé un ejemplo numérico sencillo de cada una.

**Solución**

Veamos nuestra ecuación inicial reordenada:

$$Ax^2 + Dx + Cy^2 + Ey + G = 0$$

Sacando factor común:

$$A(x^2 + \frac{D}{A}x) + C(y^2 + \frac{E}{C}y) + G = 0$$

Completando cuadrados:

$$A(x^2 + \frac{D}{A}x + \left(\frac{D}{2A}\right)^2 - \left(\frac{D}{2A}\right)^2) + C(y^2 + \frac{E}{C}y + \left(\frac{E}{2C}\right)^2 - \left(\frac{E}{2C}\right)^2) + G = 0
$$

llevandolo a un binomio cuadrado:

$$A\left( x + \frac{D}{2A} \right)^2 + C\left( y + \frac{E}{2C} \right)^2 - \left( \frac{D^2}{4A} + \frac{E^2}{4C} - G \right) = 0
$$

Dividiendo la ecuación por el termino independiente:

$$\frac{\left( x + \frac{D}{2A} \right)^2}{\left( \frac{D^2}{4A^2} + \frac{E^2}{4AC} - \frac{G}{A} \right)} + 
\frac{\left( y + \frac{E}{2C} \right)^2}{\left( \frac{D^2}{4AC} + \frac{E^2}{4C^2} - \frac{G}{C} \right)}
$$

si no fijamos, los terminos de los númeradores cumplen con x' y y'; además los del denominador coinciden con q y t, dejandonos:

$$\frac{x'^2}{q}+\frac{y'^2}{t}=1,$$

lo que cumple con la ecuación de una parabola con origen de coordendas en su centro

Condiciones para que la cónica sea una circunferencia, elipse o hipérbola:

- Circunferencia: si A=C y q=t
- Elipse: si $A\neq$ C y q>0 y t>0 
- Hipérbola: Si q>0 y t<0  o q<0  y t>0

**3.** **Rotación de cónicas.** A partir de la ecuación general de la cónica con términos acoplados

$$Ax^2 + Bxy + Cy^2 + Dx + Ey + G = 0$$

aplique una rotación de la forma

$$x  =  x' \cos\theta - y' \sin\theta $$
$$y  =  x' \sin\theta + y' \cos\theta $$

y demuestre que si se rota el sistema de coordenadas original un ángulo $\theta$ que satisface

$$\tan 2\theta = \frac{B}{A-C},$$

los nuevos ejes estarán alineados con los ejes de simetría.

**Solución**

reemplazemos 

$$x  =  x' \cos\theta - y' \sin\theta $$
$$y  =  x' \sin\theta + y' \cos\theta $$

en:

$$Ax^2 + Bxy + Cy^2 + Dx + Ey + G = 0$$

haciendo el desarrollo algebraico:

\begin{align*}
x'^2(A\cos^2\theta + B\cos\theta + C \sin^2\theta) &+ x'(D\cos\theta + E\sin\theta) \\
&+ y'^2(A\sin^2\theta - B\sin\theta \cos\theta + C \cos^2 \theta) \\
& +y'(-D\sin\theta+Ecos\theta\\
&+ x'y'(-2A\cos\theta \sin\theta + B\cos^2\theta - B\sin^2\theta) \\
&+ 2C\sin\theta cos \theta) + G = 0
\end{align*}

si queremos que loe ejes esten alineados con los ejes de simetria no deben de haber terminos cruzados, para que eso se cumpla:

\begin{align}
-2A\cos\theta + B(\cos^2\theta - \sin^2\theta) + 2C\sin\theta &= 0 \\
2A\cos\theta \sin\theta(C-A) + B(\cos^2\theta - \sin^2\theta) &= 0
\end{align}


Haciendo el algebra correspoondiente podemos llegar a que:

$$\tan 2\theta = \frac{B}{A-C}$$



**4. Anomalías.**

> a) Demuestre que, para una elipse, la distancia radial $r$ está relacionada con la anomalía excéntrica $E$ por medio de la expresión
  
$$
r=a(1 - e\cos E).
$$

> b) Demuestre, además, que la anomalía verdadera se relaciona con la anomalía excéntrica a través de
  
$$
\tan \frac{f}{2} = \sqrt{\frac{1+e}{1-e}}\tan \frac{E}{2}.
$$


**Solución**

la ecuación de una cónica respecto al centro esta dada por:

$$\frac{x_c^2}{a^2}+\frac{y_c^2}{b^2}=1$$

así la ecuación paramétrica esta dada por:

$$x_c=acosE$$
$$y_c=bsinE$$

Trasladandolo al sistema de origen en el foco podemos ver que:

$$r^2=(acosE-ae)^2+b^2sin^2E$$

por definición:

$$b^2=a^2(1-e^2)$$

Reemplazando esto en r podemos obtener que:

> $$r=a(1 - e\cos E)$$

A partir de esto se puede obtener que:

$$r cosf = a cosE - ae$$
$$r sinf = b sinE$$

y teniendo en cuenta la identidad de angulo doble:

$$tan\frac{f}{2}=\frac{1-cosf}{sinf}$$

reemplazando cosf y sinf, tenems que:

$$tan\frac{f}{2}=\frac{1-\frac{acosE-ae}{r}}{\frac{bsinE}{r}}$$

$$tan\frac{f}{2}=\frac{a(1-ecosE)-(acosE-ae)}{\sqrt{1-e^2}asinE}$$

Simplificando tenemos que:
$$\tan \frac{f}{2} = \sqrt{\frac{1+e}{1-e}}\tan \frac{E}{2}.$$

**5. Movimiento en el problema de 2 cuerpos.** El vector excentricidad está dado por $\vec{e} = \vec{A}/\mu$, donde:
  
 \begin{equation}
 \displaystyle \vec{A} = \dot{\vec{r}}\times \vec{h}  - \mu \hat{r}
 \end{equation}


> a. Haciendo el producto punto entre el vector excentricidad y el vector posición en el movimiento relativo, muestre que la trayectoria de dicho movimiento es cónica
     
> b. Usando el resultado anterior, encuentre una expresión para la distancia mínima y la distancia máxima en el caso de una órbita elíptica.
     
> c. A partir del resultado anterior, muestre que:
     
   \begin{equation}
      \frac{h^2}{\mu} = a(1-e^2)
   \end{equation}

**Solución**

Iniciando, tenemos que:

$$\vec{e} = \vec{A}/\mu$$
$$\vec{A} = \dot{\vec{r}}\times \vec{h}  - \mu \hat{r}$$

haciendo un producto punto por $\vec{r}$

$$\vec{e} \cdot \vec{r} =[\frac{\dot{\vec{r}}\times \vec{h}}{\mu}  - \hat{r}] \cdot \vec{r}$$

$$\vec{e} \cdot \vec{r} = \vec{r}\cdot\frac{\dot{\vec{r}}\times \vec{h}}{\mu}  - \hat{r} \cdot \vec{r}$$

$$ercos\theta_{er}=\frac{\vec{h}}{\mu}\cdot(\vec{r}\times \dot{\vec{r}})  - r$$

$$ercos\theta_{er}=\frac{h^2}{\mu}-r$$

despejando r tenemso que:

$$r=\frac{{h^2}/{\mu}}{1+ecos\theta_{er}}$$

La anterior es justo la expresión  de una cónica en coordenadas polares

b. veamos ahora cual es la distancia máxima y mínima en el caso de una orbita elíptica

debemos tener en cuenta que para que la función r se minimice el denominador debe de ser máximo, y al contrario para r máximo, lo que podemos variar de la ecuación es el theta, así que, rmin -> $\theta=0$  y rmax -> $\theta=\pi$

Para r min:

$$r_{min}=\frac{{h^2}/{\mu}}{1+ecos\theta_{0}}$$

$$r_{min}=\frac{{h^2}/{\mu}}{1+e}$$

Para r max:

$$r_{max}=\frac{{h^2}/{\mu}}{1+ecos\theta_{\pi}}$$

$$r_{max}=\frac{{h^2}/{\mu}}{1-e}$$  

como e para una elipse cumple que 0 < e < 1 entonces, para rmax no hay problema que el denominador sea negativo

c. veamos la relación entre h y a:

para esto debemos de tener en cuenta que rmax+rmin=2a, en el caso de la elipse:

$$r_{max}+r_{min}=2a$$

$$\frac{{h^2}/{\mu}}{1-e}+\frac{{h^2}/{\mu}}{1+e}=2a$$

$$\frac{{h^2}/{\mu}(1+e+1-e)}{(1-e^2)}=2a$$

así tenemos que:

$$\frac{h^2}{\mu} = a(1-e^2)$$

**6. Momentum angular orbital de un sistema binario.** Dos masas $m_1$ y $m_2$ forman un sistema estelar binario describiendo órbitas elípticas alrededor de su centro de masa común con un periodo $P$. La mayor separación entre ambas estrellas es $D_1$ y la menor $D_2$. Muestre que la magnitud del momentum angular orbital del sistema es:

$$
\displaystyle L = \frac{\pi}{P}\frac{m_1m_2}{m_1+m_2}\sqrt{D_1D_2}(D_1+D_2)
$$

**Solución**

Para esto debemos tener en cuenta un resultado del problem set 1:

$$\vec{L} = M\vec{R}\times \dot{\vec{R}} + m_r \vec{r}\times \dot{\vec{r}}$$

si definimos el origen de nuestro sistema coordenado en el centro de masas, tenemos que:

$$\vec{L} =  m_r \vec{r}\times \dot{\vec{r}}$$

así:

$$L=mrh$$

teniendo el cuenta el resultado del punto 5 $\frac{h^2}{\mu} = a(1-e^2)$, despejando h y reemplazando en L:

$$L=m_r\sqrt{\mu a(1-e^2)}$$

Ahora expresemos $(1-e^2)$ en terminos de a y D:

$$r_{min}=D_1=\frac{{h^2}/{\mu}}{1+e}=\frac{a(1-e^2)}{1+e}=a(1-e)$$

$$r_{max}=D_2=\frac{{h^2}/{\mu}}{1-e}=\frac{a(1-e^2)}{1-e}=a(1+e)$$

$$D_1 D_2=a^2(1-e)$$

reemplazando esto en la ultima expresión para L:

$$L=m_r\sqrt{\mu \frac{D_1 D_2}{a}}$$

si tenemos en cuenta que la terecera ley de Kepler nos dice que $\mu=\frac{a^3 4\pi^2}{P^2}$, lo podemos reemplazar, y nos queda:

$$L=m_r\sqrt{\frac{a^3 4\pi^2}{P^2} \frac{D_1 D_2}{a}}$$

$$L=m_r\sqrt{\frac{a^2 4\pi^2}{P^2} D_1 D_2}$$

sabemos que $a^2=\frac{(D_1+D_2)^2}{4}$, podemos reemplazarlo:


$$L=m_r\sqrt{\frac{(D_1+D_2)^2}{4} \frac{ 4\pi^2}{P^2} D_1 D_2}$$


de aqui se puede deducir facilmente con un poco de algebra que:


$$L = \frac{\pi}{P}\frac{m_1m_2}{m_1+m_2}\sqrt{D_1D_2}(D_1+D_2)$$

**7. Parámetro de impacto.** Considere un cometa en una orbita hiperbólica con el Sol en su foco. El parámetro de impacto, $b$, se define como la distancia de acercamiento más cercana en ausencia de atracción gravitacional entre el cometa y el Sol. Demuestre que

\begin{equation}
b = \frac{h}{\sqrt{2ε}}
\end{equation}

donde $h$ y $ε$ son el momentum angular y la energía específicos, respectivamente. Muestre que la relación entre el parámetro de impacto y el verdadero punto más cercano al Sol, $r_p$, es

\begin{equation}
r_p = \frac{2b}{\alpha + \sqrt{\alpha^2 + 4}}
\end{equation}

donde $\alpha = GM/(ε b)$ y $M$ es la masa solar. Finalmente, deduzca que si el cometa no colisiona con el Sol, entonces:

\begin{equation}
\alpha < \frac{b}{R} - \frac{R}{b}
\end{equation}

**Solución:**

sabemos que:

$$\vec{h}=\vec{r} X \dot{\vec{r}}$$

$$\epsilon=\frac{1}{2}v^2-\frac{\mu}{r}$$

En el infinito:

$$v=v_{\infin}$$
$$r->\infin$$

así que:

$$\epsilon=\frac{1}{2}v^2$$

dejandonos que:

$$v_{\infin}=\sqrt{2\epsilon}$$

si separamos la posición del cuerpo en su componente paralela  y prependicular, donde definimos la distancia paralela como la que va en la dirección de la trayectoria del cuerpo y la permedicula como la que va en la dirección de la distancia mínima entre el cuerpo y el sol, así:

$$ \vec{h} = ( \vec{r}_{\perp} + \vec{r}_{\parallel}) X \vec{v}$$

la componente paralela se cancela  en el punto más cercano al sol en ausencia de gravedad, y la perpendicular nos representa el paramétro b:

$$\vec{h} =\vec{r}_{\perp}X \vec{v}$$

en este punto r y v son paralelos así que:

$$h=bv_{\infin}$$

$$h=b\sqrt{2\epsilon}$$

$$b=\frac{h}{\sqrt{2\epsilon}}$$

Para la siguiente demostración, tenemos que:

$$\alpha = GM/(ε b) $$

como la masa del cometa es muy pequeña comparada con la del sol:

$$\alpha = \mu /(ε_0)$$

Ahora:

$$r=\frac{\frac{h^2}{\mu}}{1+ecos\theta}$$

$$r_p=r(0)=\frac{\frac{h^2}{\mu}}{1+e}$$

la excentricidad esta definida como:

$$e=\sqrt{1+\frac{2\epsilon h^2}{\mu^2}}$$

realizando las operaciones algebraicas correctas nos podemos dar de cuenta que:

$$r_p=\frac{2b}{\alpha+\sqrt{\alpha^2+4}}$$

debemos tener en cuenta que $r_p>R$ donde R es el radio del sol, porque si no el cometa chocaria con el sol sin importar el valor de b

$$\frac{2b}{\alpha+\sqrt{\alpha^2+4}}>R$$

realizando los correctos procedimientos algebraicos, tenemos que:

$$\alpha< \frac{b}{R}-\frac{R}{b}$$

$$l=M-f$$

$$\frac{\partial M}{\partial E}-\frac{\partial f}{\partial E}=0$$

$$M=E-einE$$
$$\frac{\partial M}{\partial E}=1-ecosE$$

Ahora, para f:

$$tan \frac{f}{2}=\sqrt{\frac{1+e}{1-e}}tan\frac{E}{2}$$

$$1-ecosE-\frac{\sqrt{1-e^2}}{1-ecosE}=0$$

con algebra básica se obtiene que:

$$cosE=\frac{1-\sqrt[4]{1-e^2}}{e}$$

sabemos que r=a(1-ecosE), despejando ecosE, y uniendo ambas ecuaciones podemos obtener que:

> $$r=a+\sqrt[4]{1-e^2}$$
